```python
#making a dataframe
def PlaystoreInfo():
  PSinfo_df = pd.DataFrame(index=play_store_dataset1.columns)
  PSinfo_df['Datatypes'] =  play_store_dataset1.dtypes
  PSinfo_df['Count of non-null values'] = play_store_dataset1.count()
  PSinfo_df['NaN values'] = play_store_dataset1.isnull().sum()
  PSinfo_df['% NaN Values'] = (PSinfo_df['NaN values']/len(play_store_dataset1)).round(4)*100 # or play_store_dataset1.isnull().mean()
  PSinfo_df['Unique_count'] = play_store_dataset1.nunique()
  return PSinfo_df
PlaystoreInfo() 
```

##**Size Column Operation**

```python
#Convert all size to KB so Each M have to Multiply with 1024
#Removing "M", Changing Size To KB
play_store_dataset2['Size'] = play_store_dataset2['Size'].map(
    lambda value :
        str(int(float(value.rstrip('M')) * 1024)) if value[-1] == 'M' else value
)

# Removing "k"
play_store_dataset2['Size'] = play_store_dataset2['Size'].map(
    lambda value :
        str(value.rstrip('k')) if value[-1] == 'k' else value
)

# Setting "Varies with device" to NaN
play_store_dataset2['Size'] = play_store_dataset2['Size'].map(
    lambda value :
        np.nan if value == 'Varies with device' else value
)
```
